```python
#Challenge 1
bins = [0, 1000, 10000, 100000, 1000000, float('inf')]
labels = ['cat_1', 'cat_2', 'cat_3', 'cat_4', 'cat_5']
movie_data['bin'] = pd.cut(movie_data['Votes'], bins=bins, labels=labels, right=False)
movie_data
```

```python
movie_data.columns
```

```python
#Challenge 2
movie_data['revenue_per_minute'] = movie_data['Revenue (Millions)'] / movie_data['Runtime (Minutes)']
movie_data
```

```python
#Challenge 3
movie_data['new_rating'] = 0
movie_data.loc[movie_data['Genre'] == 'thriller', 'new_rating'] += 1
movie_data.loc[movie_data['Genre'] == 'comedy', 'new_rating'] -= 1
movie_data
```

```python
#Challenge 4
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True
movie_data['is_prime'] = movie_data.apply(lambda row: is_prime(int(sum(ord(char) for char in row['Title']) / row['Votes'])), axis=1)
```

```python
movie_data
```

```python
#Challenge 5
movie_data['popularity_score'] = movie_data['Votes'] + movie_data['Runtime (Minutes)'] + movie_data['Revenue (Millions)']
```

```python
movies_ranked = movie_data.sort_values(by='popularity_score', ascending=False)
movies_ranked
```

```python
#Bonus Challenge

import hashlib

def sum_numeric_values(s):
    return sum(int(char) for char in s if char.isdigit())

movie_data['description_hash'] = movie_data['Description'].apply(lambda x: hashlib.sha256(x.encode()).hexdigest())


movie_data['hash_numeric_sum'] = movie_data['description_hash'].apply(sum_numeric_values)


movie_data['hidden_pattern'] = (movie_data['hash_numeric_sum'] > movie_data['Revenue (Millions)']) & (movie_data['hash_numeric_sum'] < movie_data['Votes'])


movies_with_hidden_patterns = movie_data[movie_data['hidden_pattern']]


print(movies_with_hidden_patterns)
```
